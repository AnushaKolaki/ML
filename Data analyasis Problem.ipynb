# Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
dataset["Black"]=le.fit_transform(dataset["Black"])
dataset["Event"]=le.fit_transform(dataset["Event"])
dataset["Opening"]=le.fit_transform(dataset["Opening"])
dataset["Result"]=le.fit_transform(dataset["Result"])
dataset["Site"]=le.fit_transform(dataset["Site"])
dataset["White"]=le.fit_transform(dataset["White"])
dataset["ECO"]=le.fit_transform(dataset["ECO"])

dataset
```

```python
x=dataset.iloc[:,[0,1,2,4,5,7,8,9,11,12,13]].values
x
```

```python
y=dataset.iloc[:,[11]].values
y
```

# Split for train & Test model

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=11)
```

# SVM

```python
from sklearn.svm import SVC
sv=SVC(kernel="linear",random_state=11)
sv.fit(x,y)
```

```python
y_pred=sv.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
ac=accuracy_score(y_pred,y_test)*100
ac
```
